```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
%matplotlib inline
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
df.info()
df.describe()
df.isnull().sum()
plt.figure(figsize=(8,5))
df['Age'].hist(bins=30, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
df.isnull().sum()
plt.figure(figsize=(8,5))
df['Age'].hist(bins=30, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Boxplot of Age by Survival')
plt.show()
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
plt.title('Boxplot of Age by Survival')
plt.show()
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
sns.pairplot(df[['Age', 'Fare', 'Survived']])
plt.show()
df.groupby('Sex')['Survived'].mean()
plt.title('Correlation Matrix')
plt.show()
sns.pairplot(df[['Age', 'Fare', 'Survived']])
plt.show()
df.groupby('Sex')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,20,40,60,80], labels=['Child','Teen','Adult','Middle-aged','Senior'])
df.groupby('AgeGroup')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,20,40,60,80], labels=['Child','Teen','Adult','Middle-aged','Senior'])
df.groupby('AgeGroup')['Survived'].mean()
df['Age'].fillna(df['Age'].median(), inplace=True)
df.drop(columns=['Cabin'], inplace=True)
df.dropna(inplace=True)
df.to_csv('cleaned_titanic.csv', index=False)
files.download('cleaned_titanic.csv')
```
